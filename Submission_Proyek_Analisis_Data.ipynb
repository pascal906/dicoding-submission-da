### Cleaning Data

* karena tidak ada missing value, invalid value, inaccurate value, duplicate value, dan juga outlier masih bisa dimaklumi
* melakukan perbaikan kecil pada nama kolom
* melakukan pengubahan nilai kategorik yang numerik ke string

#### perbaikan nama kolom

```python
# melihat nama kolom dari dataset
hour_df.columns
```

```python
# mengganti nama kolom
hour_df = hour_df.rename(columns={
    'dteday': 'date',
    'yr': 'year',
    'mnth': 'month',
    'hr': 'hour',
    'cnt': 'count',
})
hour_df.head()
```

#### mapping nilai kategori numerik

```python
# mapping nilai kategori numerik
season_map = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'}
year_map = {0: '2011', 1: '2012'}
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
holiday_map = {0: 'No', 1: 'Yes'}
weekday_map = {0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'}
workingday_map = {0: 'No', 1: 'Yes'}
weathersit_map = {1: 'Clear', 2: 'Mist', 3: 'Light Rain', 4: 'Heavy Rain'}

hour_df['season'] = hour_df['season'].map(season_map)
hour_df['year'] = hour_df['year'].map(year_map)
hour_df['month'] = hour_df['month'].map(month_map)
hour_df['holiday'] = hour_df['holiday'].map(holiday_map)
hour_df['weekday'] = hour_df['weekday'].map(weekday_map)
hour_df['workingday'] = hour_df['workingday'].map(workingday_map)
hour_df['weathersit'] = hour_df['weathersit'].map(weathersit_map)

hour_df.head()
```

#### mengganti tipe data kolom

```python
# konversi tipe data kategori
order_season = ['Winter', 'Spring', 'Summer', 'Fall']
order_year = ['2011', '2012']
order_month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
order_holiday = ['No', 'Yes']
order_weekday = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
order_workingday = ['No', 'Yes']
order_weathersit = ['Clear', 'Mist', 'Light Rain', 'Heavy Rain']

hour_df['season'] = pd.Categorical(hour_df['season'], categories=order_season, ordered=True)
hour_df['year'] = pd.Categorical(hour_df['year'], categories=order_year, ordered=True)
hour_df['month'] = pd.Categorical(hour_df['month'], categories=order_month, ordered=True)
hour_df['holiday'] = pd.Categorical(hour_df['holiday'], categories=order_holiday, ordered=True)
hour_df['weekday'] = pd.Categorical(hour_df['weekday'], categories=order_weekday, ordered=True)
hour_df['workingday'] = pd.Categorical(hour_df['workingday'], categories=order_workingday, ordered=True)
hour_df['weathersit'] = pd.Categorical(hour_df['weathersit'], categories=order_weathersit, ordered=True)

# mengubah tipe data dteday jadi datetime
hour_df['date'] = pd.to_datetime(hour_df['date'], errors='coerce')

hour_df.info()
```

Data hour_df yang telah kita bersihkan tadi, bisa kita gunakan juga sebagai day_df yang sebelumnya, tinggal diubah ke bentuk hari.

```python
# membuat dataset day_df dari dataset hour_df yang telah dibersihkan
day_df = hour_df.groupby(by='date').agg({
    'season': 'first',
    'year': 'first',
    'month': 'first',
    'holiday': 'first',
    'weekday': 'first',
    'workingday': 'first',
    'weathersit': 'first',
    'casual': 'sum',
    'registered': 'sum',
    'count': 'sum'
}).reset_index()
day_df.head()
```

```python
# informasi dataset day_df
day_df.info()
```

**Insight:**

* data sudah bersih dan siap digunakan
* beberapa perbaikan pada nama kolom membuat data lebih mudah dipahami
* setelah data sudah bersih, karena hanya data hour yang kita bersihkan, maka selanjutnya kita buat data hour ini jadi day untuk digunakan dalam analisis nantinya

##Analysis Lanjutan

```python
# recency
last_date = day_df['date'].max()
day_df['recency'] = (last_date - day_df['date']).dt.days

# hari dengan share tertinggi treshold(5000)
high_share_day = day_df[day_df['count'] > 5000]

last_high_share = high_share_day['date'].max()
recency_score = (last_date - last_high_share).days
print("Recency Score:", recency_score)

# frequency score
frequency_score = high_share_day.shape[0]
print("Frequency Score:", frequency_score)

# monetary total peminjaman sepeda selama hari high-rent
monetary_score = high_share_day['count'].sum()
print("Monetary Score:", monetary_score)
```

dari sini diketahui bahwa:
* Recency Score: pada 12 hari terakhir pada data terjadi peningkatan peminjaman sepeda
* Frequency Score: pada data per hari, terdapat 286 hari dimana peminjamannya lebih dari 5000
* Monetary Score: menunjukkan pada data, selama 286 hari peminjaman terbanyak itu, total peminjaman mencapai 1,846,724 peminjaman.
